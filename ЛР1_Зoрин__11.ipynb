### Задание №1.
Нужно выделить минимум 3 категории (маленький, большой, средний), фильтрацию для уровня пробега выбрать самостоятельно, аргументировать выбор и создать сводную таблицу: средний и медианный пробег по категории пробега и типу топлива.

Ход выполнения задания:

1) Используем pd.qcut() для разделения пробега на 3 квантильные категории.

2) Строим pivot_table, где:индекс — категория пробега (kmd_cat); столбцы — вид топлива (Fuel); значения — пробег (Kmdriven); агрегаты — среднее и медиана.

3) Получаем таблицу, отражающую различия по топливу и категории пробега.

```python
df['kmd_cat'] = pd.qcut(df['kmdriven'], q=3, labels=['Small', 'Medium', 'Large'])

add_pivot = df.pivot_table(
    values='kmdriven',
    index='kmd_cat',
    columns='fuel',
    aggfunc=['mean', 'median'],
    observed=False
).round(2)

print("Задание 1 — сводная таблица (mean и median Kmdriven) по категории пробега и типу топлива:")
display(add_pivot)
```

В данном задании используются равные по частоте категории (qcut) — это обоснованный выбор, так как он делит данные по фактическому распределению пробега. Результаты показали чёткие зависимости между типом топлива и категориями пробега. У автомобилей с дизельным двигателем средний и медианный пробег значительно выше, чем у бензиновых и газовых машин. Это объясняется тем, что дизельные авто чаще используются для дальних поездок, грузоперевозок и служебных нужд, где важна экономичность. Электромобили и машины на газе демонстрируют самый низкий пробег, что связано с их меньшей распространённостью и недавним появлением на рынке. У бензиновых машин показатели средние — они преобладают среди личных автомобилей и чаще используются для повседневных поездок. Наблюдаемые различия в пробеге помогают оценить, как тип топлива влияет на срок эксплуатации и потенциальную износостойкость авто. Такой анализ полезен для прогнозирования остаточного ресурса и определения рыночной стоимости подержанных машин разных типов топлива.

Нужно создать столбец “Категория стоимости (price)” (с помощью категоризации).Далее выделить минимум 3 категории (низкая, высокая, средняя), фильтрацию для уровня цены выбрать самостоятельно, аргументировать выбор. И затем создать сводную таблицу: средняя и медианная price по категории цены и году.

Ход выполнения работы:

1) Разбиваем SellingPrice на три категории с помощью pd.qcut().

3) Анализируем, как стоимость зависит от возраста автомобиля.

Средние по цене автомобили занимают промежуточное положение — чаще всего это модели 2013–2017 годов, находящиеся в активной эксплуатации. Медианные значения почти всегда меньше средних, что говорит о наличии отдельных дорогих экземпляров, повышающих среднее значение. ## Задание №16

Необходимо создать столбец “Категория года (year)” (с помощью категоризации)7. Далее выделить минимум 3 категории (старая, новая, средняя), фильтрацию для года выбрать самостоятельно, аргументировать выбор. И затем создать группировку: средний, максимальный, минимальный и медианный пробег по категории года.

Ход выполнения задания:
1) Делим автомобили по возрасту с помощью pd.qcut() по столбцу year.

2) Применяем groupby('year_cat')['kmdriven'].agg(['mean','max','min','median']).

3) Получаем агрегированные данные, описывающие распределение пробега в каждой возрастной категории.

```python
df['year_cat'] = pd.qcut(df['year'], q=3, labels=['Old', 'Mid', 'New'])

add_pivot_3 = df.groupby('year_cat', observed=False)['kmdriven'].agg(['mean', 'max', 'min', 'median']).round(2)

print("Задание 16 — агрегаты Kmdriven по категории года (mean, max, min, median):")
display(add_pivot_3)
```

## Задание №25

В задании 25 нужно создать столбец “Категория стоимости (price)” (с помощью категоризации).
Выделить минимум 3 категории (низкая, высокая, средняя), фильтрацию для
уровня цены выбрать самостоятельно, аргументировать выбор. Отфильтровать
набор данных. Выбрать только записи, где категория цены= средняя+ выбрать
2 типа топлива с минимальной медианной стоимостью + transmission c
максимальным количеством записей + owner с максимальным количеством
записей. Создать группировки : средний, максимальный, минимальный,
медианный пробег и средняя, максимальная, минимальная, медианная цена по
году и по типу seller_type.

Ход выполнения задания:

1) Создаём price_cat (Low, Medium, High).

2) Берём только машины категории “Medium”.

3) Находим 2 типа топлива с минимальной медианной ценой (fuel_medians.sort_values().head(2)),
самую популярную трансмиссию (value_counts().idxmax()),
и самого частого владельца (value_counts().idxmax()).

4) Фильтруем DataFrame по этим условиям.

5) Строим pivot_table с агрегатами (mean, median, max, min) для SellingPrice и kmdriven по year и seller_type.

```python
df['price_cat'] = pd.qcut(df['SellingPrice'], q=3, labels=['Low', 'Medium', 'High'])

df_mid = df[df['price_cat'] == 'Medium']

fuel_medians = df_mid.groupby('fuel')['SellingPrice'].median().sort_values()
top2_fuels = fuel_medians.head(2).index.tolist()

top_trans = df_mid['transmission'].value_counts().idxmax()

top_owner = df_mid['owner'].value_counts().idxmax()

df_filtered = df_mid[
    (df_mid['fuel'].isin(top2_fuels)) &
    (df_mid['transmission'] == top_trans) &
    (df_mid['owner'] == top_owner)
]

pivot_25 = df_filtered.pivot_table(
    values=['kmdriven', 'SellingPrice'],
    index=['year', 'seller_Type'],
    aggfunc=['mean', 'median', 'max', 'min']
).round(2)

print("Задание 25 — сводная таблица (по году и типу продавца):")
display(pivot_25)
```

В задании 25 выполнялась сложная фильтрация: были выбраны автомобили со средней ценой, два вида топлива с минимальной медианной стоимостью, популярная трансмиссия и тип владельца. Далее по этим данным были построены группировки по году и типу продавца. Это упражнение объединяет все ключевые приёмы анализа данных: категоризацию, фильтрацию, группировку и сводные таблицы. Итоговый набор показывает поведение машин среднего ценового сегмента. Выбор топовых видов топлива, трансмиссии и владельцев позволяет сосредоточиться на наиболее массовых категориях рынка. Анализ показывает, что бензиновые автомобили с механикой, проданные частными владельцами, составляют основную часть выборки и демонстрируют сбалансированные показатели цены и пробега. У дилеров обычно цены выше, но пробеги меньше — это типичная рыночная закономерность. Такое исследование помогает понять, какие комбинации характеристик наиболее типичны и выгодны, и может служить основой для рекомендаций при оценке, покупке или прогнозировании стоимости подержанных автомобилей.
